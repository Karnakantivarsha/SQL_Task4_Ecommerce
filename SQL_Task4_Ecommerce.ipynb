```python
import pandas as pd
import sqlite3
```

```python
df = pd.read_excel('/content/Online Retail.xlsx')
df.head()
```

```python
# Create SQLite database connection
conn = sqlite3.connect('ecommerce.db')

# Load dataframe into SQL table
df.to_sql('ecommerce', conn, index=False, if_exists='replace')

print("✅ Database created and table 'ecommerce' loaded successfully!")
```

```python
pd.read_sql_query("SELECT * FROM ecommerce LIMIT 5;", conn)
```

```python
query1 = "SELECT * FROM ecommerce LIMIT 10;"
pd.read_sql_query(query1, conn)
```

```python
query2 = """
SELECT CustomerID, SUM(Quantity * UnitPrice) AS TotalRevenue
FROM ecommerce
WHERE CustomerID IS NOT NULL
GROUP BY CustomerID
ORDER BY TotalRevenue DESC;
"""
pd.read_sql_query(query2, conn)
```

```python
query3 = """
SELECT Country, SUM(Quantity * UnitPrice) AS TotalRevenue
FROM ecommerce
GROUP BY Country
ORDER BY TotalRevenue DESC
LIMIT 5;
"""
pd.read_sql_query(query3, conn)
```

```python
query4 = """
SELECT CustomerID, AVG(Quantity * UnitPrice) AS AvgOrderValue
FROM ecommerce
WHERE CustomerID IS NOT NULL
GROUP BY CustomerID
ORDER BY AvgOrderValue DESC
LIMIT 10;
"""
pd.read_sql_query(query4, conn)
```

```python
query5 = """
SELECT
    strftime('%Y-%m', InvoiceDate) AS Month,
    SUM(Quantity * UnitPrice) AS MonthlyRevenue
FROM ecommerce
GROUP BY Month
ORDER BY Month;
"""
pd.read_sql_query(query5, conn)
```

```python
query6 = """
SELECT Description, SUM(Quantity) AS TotalQuantitySold
FROM ecommerce
GROUP BY Description
ORDER BY TotalQuantitySold DESC
LIMIT 10;
"""
pd.read_sql_query(query6, conn)
```

```python
df2 = pd.read_sql_query(query2, conn)
df2.to_csv('total_revenue_per_customer.csv', index=False)
```
